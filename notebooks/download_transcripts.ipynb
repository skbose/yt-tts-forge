```python
import pytube
```

```python
akshat_video = "https://www.youtube.com/watch?v=UGpqSP2P6Vw&t=51s&ab_channel=AkshatShrivastava"
```

```python
yt = pytube.YouTube(akshat_video)
```

```python
stream = yt.streams.filter(only_audio=True).first()
stream.download("output")
```

```python
yt.captions
```

```python
caption = yt.captions.get_by_language_code('en-IN')
caption_text = caption.generate_srt_captions()
```

```python
# write the caption to a file
with open("output/caption.srt", "w") as file:
    file.write(caption_text)
```

```python
# parse the srt file as timestamp and text
with open("output/caption.srt", "r") as file:
    lines = file.readlines()
    lines = [line.strip() for line in lines]
    print (lines)
```
